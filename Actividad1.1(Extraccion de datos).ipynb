```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df.head()
```

```python
df.info()
```

### Filtro 1
Este filtro nos regresa los registros en los cuales su valor para la columna ***'NOMBRE_CLIENTE'*** es igual a ***'JUAN ANTONIO RESENDIZ UGALDE'***.

```python
# Filtro por objeto
filtro1 = df[df['NOMBRE_CLIENTE'] == 'JUAN ANTONIO RESENDIZ UGALDE']
filtro1.to_csv('Filtro01.csv')
filtro1
```

### Filtro 2
En este filtro se estarán regresando únicamente las filas o registros con índice ***45, 1050 y 1200***.

```python
#filtro por filas
filtro2 = df.iloc[[45,1050,1200],:]
filtro2.to_csv('Filtro02.csv')
filtro2
```

### Filtro 3
En este filtro se regresan todos los registros mostrando únicamente las columnas ***1, 4 y 5***.

```python
#Filtro por columnas
filtro3 = df.iloc[:,[1,4,5]]
filtro3.to_csv('Filtro03.csv')
filtro3
```

### Filtro 4
Primero creamos un nuevo DataFrame indicando que la columna 2 (***'NOMBRE_VENDEDOR'***) sea el nuevo índice. El filtro, nos regresa la columna ***'MARGEN_TOTAL_CALCULADO'***  con los registros con índices ***'LETICIA RAMIREZ HERNANDEZ' y 'AARON NUÑO CORDOVA'***

```python
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
filtro4 = df1.loc[['LETICIA RAMIREZ HERNANDEZ', 'AARON NUÑO CORDOVA'],['MARGEN_TOTAL_CALCULADO']]
filtro4.to_csv('Filtro04.csv')
filtro4
```

### Filtro 5
Este filtro usa la función ***.head()*** para regresar los primeros 5 registros del DataFrame.

```python
filtro5 = df.head()
filtro5.to_csv('Filtro05.csv')
filtro5
```

### Filtro 6
Este filtro usa la función ***.tail()*** para regresar los últimos 5 registros del DataFrame.

```python
filtro6 = df.tail()
filtro6.to_csv('Filtro06.csv')
filtro6
```

### Filtro 7
Este filtro nos regresa los registros para los cuales su valor en la columna ***'MARGEN_TOTAL_CALCULADO'*** es mayor a ***7000***

```python
# Filtro de comparación
filtro7 = df1[df1['MARGEN_TOTAL_CALCULADO'] > 7000]
filtro7.to_csv('Filtro07.csv')
filtro7
```

### Filtro 8
Este filtro nos regresa los registros para los cuales su valor en la columna ***'MARGEN_TOTAL_CALCULADO'*** es mayor a ***7000*** y su valor en la columna ***'FECHA_DOC'*** es igual a ***'2022-12-23'***

```python
#Filtro Y
filtro8 = df[(df['MARGEN_TOTAL_CALCULADO'] > 7000) & (df['FECHA_DOC'] == '2022-12-23')]
filtro8.to_csv('Filtro08.csv')
filtro8
```

### Filtro 9
Este filtro nos regresa los registros para los cuales su valor en la columna ***'MARGEN_TOTAL_CALCULADO'*** es mayor a ***7000*** ó su valor en la columna ***'FECHA_DOC'*** es igual a ***'2022-12-23'***

```python
#Filtro o
filtro9 = df[(df['MARGEN_TOTAL_CALCULADO'] > 7000) | (df['FECHA_DOC'] == '2022-12-23')]
filtro9.to_csv('Filtro09.csv')
filtro9
```
